## Adding titles

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
a = sns.load_dataset("iris")
a

#draw a line plot

sns.lineplot( x="sepal_length", y="sepal_width", data=a)
plt.title("phoolon ka plot")
plt.show()

```

## Adding limits

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
a = sns.load_dataset("iris")
a

#draw a line plot

sns.lineplot( x="sepal_length", y="sepal_width", data=a)
plt.xlim(2)
plt.ylim(1)
plt.show()

```

## Set styles
- drakgrid
- whitegrid
- dark
- white
- ticks

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style(style=None, rc=None)
# load dataset
a = sns.load_dataset("iris")
a

#draw a line plot

sns.lineplot( x="sepal_length", y="sepal_width", data=a)
sns.set_style(style="dark")
plt.show()

```

## Size of figure

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
a = sns.load_dataset("iris")
a
#figure size
plt.figure(figsize=(10,6))

#draw a line plot

sns.lineplot( x="sepal_length", y="sepal_width", data=a)
plt.show()

```

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
a = sns.load_dataset("iris")
a

#draw a line plot

sns.lineplot( x="sepal_length", y="sepal_width", data=a)
# set style
sns.set_style("white")
sns.set_style(style=None, rc=None)
plt.show()
```
